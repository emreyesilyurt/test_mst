```python

    import numpy as np
    def write_to_supabase(self, item):
        """
        Write processed data to Supabase PostgreSQL database using repository classes.
        If connection fails due to network issues, log the data instead.
        This method uses synchronous database operations to avoid async context issues.

        Args:
            data: The data to write to Supabase.
        """
        try:
            if self.session is None:
                print("Initializing database session...")
                from sqlalchemy.orm import Session
                from src.db.models import engine

                engine = engine
                self.session = Session(engine)
                print(f"Session initialized: {self.session}")

            # Use repository for product data - assuming repositories can work with sync session
            print("Setting up repository objects for database operations...")
            product_repo = ProductRepository(self.session)
            manufacturer_repo = ManufacturerRepository(self.session)
            part_number_repo = PartNumberRepository(self.session)
            doc_media_repo = DocumentMediaRepository(self.session)
            attr_repo = ProductAttributeRepository(self.session)
            extra_repo = ProductExtraRepository(self.session)
            print("Repository objects created successfully.")


            if not isinstance(item, dict):
                return {"status": "error", "message": f"Skipping item: {item} - Not a dictionary"}

            # Process manufacturer
            manufacturer_name = item.get('manufacturer', '')
            existing_manufacturers = []
            manufacturer = None
            if hasattr(manufacturer_repo, 'list_sync'):
                existing_manufacturers = manufacturer_repo.list_sync(name=manufacturer_name)
            else:
                # Fallback to a direct SQLAlchemy query if possible
                from sqlalchemy import select
                stmt = select(Manufacturer).where(Manufacturer.name == manufacturer_name)
                result = self.session.execute(stmt)
                existing_manufacturers = result.scalars().all()
            manufacturer = existing_manufacturers[0] if existing_manufacturers else None

            if not manufacturer:
                # Create a new manufacturer record, assuming it's normalized for now
                manufacturer = Manufacturer(name=manufacturer_name)
                self.session.add(manufacturer)
                self.session.flush()
                # Update manufacturer_id to be the same as id for new records (assuming normalized)
                manufacturer.manufacturer_id = manufacturer.id
                self.session.commit()  # Commit after manufacturer update
            elif manufacturer.manufacturer_id != manufacturer.id:
                # If manufacturer_id is different, use the normalized record's id
                from sqlalchemy import select
                stmt = select(Manufacturer).where(Manufacturer.id == manufacturer.manufacturer_id)
                result = self.session.execute(stmt)
                normalized_manufacturer = result.scalars().first()
                if normalized_manufacturer:
                    manufacturer = normalized_manufacturer

            # Process part number
            part_number_name = item.get('pn', '')
            existing_part_numbers = []
            part_number = None
            if hasattr(part_number_repo, 'list_sync'):
                existing_part_numbers = part_number_repo.list_sync(name=part_number_name)
            else:
                from sqlalchemy import select
                stmt = select(PartNumber).where(PartNumber.name == part_number_name)
                result = self.session.execute(stmt)
                existing_part_numbers = result.scalars().all()
            part_number = existing_part_numbers[0] if existing_part_numbers else None

            if not part_number:
                # Create new product first to get product_id
                part_number = PartNumber(name=part_number_name)
                self.session.add(part_number)
                self.session.flush()  # Assign product_id

                part_number.product_id = part_number.id
                self.session.commit()  # Commit after manufacturer update
            elif part_number.product_id != part_number.id:
                # If manufacturer_id is different, use the normalized record's id
                from sqlalchemy import select
                stmt = select(PartNumber).where(PartNumber.id == PartNumber.product_id)
                result = self.session.execute(stmt)
                normalized_partnumber = result.scalars().first()
                if normalized_partnumber:
                    part_number = normalized_partnumber


            product = None
            from sqlalchemy import select, and_
            stmt = select(Product).where(
                and_(
                    Product.part_number == (part_number.product_id if part_number and part_number.product_id == part_number.id else (part_number.product_id if part_number else 0)),
                    Product.manufacturer_id == (manufacturer.manufacturer_id if manufacturer and manufacturer.manufacturer_id == manufacturer.id else (manufacturer.manufacturer_id if manufacturer else 0))
                )
            )
            result = self.session.execute(stmt)
            existing_product = result.scalars().first()
            if existing_product:
                product = existing_product
            else:
                product = Product(
                    part_number=part_number.product_id if part_number and part_number.product_id == part_number.id else (part_number.product_id if part_number else 0),
                    manufacturer_id=manufacturer.manufacturer_id if manufacturer and manufacturer.manufacturer_id == manufacturer.id else (manufacturer.manufacturer_id if manufacturer else 0)
                )
                self.session.add(product)
                self.session.flush()  # Flush to assign product_id before commit
                self.session.commit()

                # Create or fetch Category
                category_names = item.get('category')
                if category_names:
                    category_name = category_names.split('->')[-1].strip()
                else:
                    category_name = 'Uncategorized'
                from sqlalchemy import select
                from src.db.models.models import Category
                stmt = select(Category).where(Category.name == category_name)
                result = self.session.execute(stmt)
                category = result.scalars().first()
                if not category:
                    category = Category(name=category_name)
                    self.session.add(category)
                    self.session.flush()

                # Set product category_id if available
                if hasattr(product, 'category_id'):
                    self.session.add(product)  # Ensure product is added to session
                    product.category_id = category.id
                    self.session.flush()
                    self.session.commit()

            # Process additional data like attributes, media, etc. if available
            if 'specs' in item and isinstance(item['specs'],np.ndarray) and item['specs'].size > 0:
                for attr in item['specs']:
                    attr_type = attr.get('spec_name', 'Unknown')

                  # Create or fetch Attribute
                    attribute = None
                    from sqlalchemy import select
                    from src.db.models.models import Attribute
                    stmt = select(Attribute).where(Attribute.name == attr_type)
                    result = self.session.execute(stmt)
                    attribute = result.scalars().first()
                    if not attribute:
                        attribute = Attribute(name=attr_type)
                        self.session.add(attribute)
                        self.session.flush()

                    # Establish relationship in CategoryAttribute if both exist
                    if category and attribute:
                        from sqlalchemy import select
                        from src.db.models.models import CategoryAttribute
                        stmt = select(CategoryAttribute).where(
                            CategoryAttribute.category_id == category.id,
                            CategoryAttribute.attribute_id == attribute.id
                        )
                        result = self.session.execute(stmt)
                        cat_attr = result.scalars().first()
                        if not cat_attr:
                            cat_attr = CategoryAttribute(
                                category_id=category.id,
                                attribute_id=attribute.id
                            )
                            self.session.add(cat_attr)
                            self.session.commit()

                    # Add ProductAttribute with the correct attribute_id if attribute exists
                    if attribute:
                        print(f"Processing product attribute: type={attr_type}, product_id={product.product_id if hasattr(product, 'product_id') else 0}, attribute_id={attribute.id}")
                        attr_obj = ProductAttribute(
                            product_id=product.product_id if hasattr(product, 'product_id') else 0,
                            attribute_id=attribute.id,
                            value_text=attr.get('spec_value', ''),
                            value_units=attr.get('spec_units', ''),
                            source=self.source,
                            ref_id=self.ref_id  # Use the ref_id from impute_op
                        )
                        self.session.add(attr_obj)
                        print(f"Added product attribute for product_id {product.product_id if hasattr(product, 'product_id') else 0} with type: {attr_type} and attribute_id: {attribute.id}")
                    else:
                        print(f"Skipping product attribute for product_id {product.product_id if hasattr(product, 'product_id') else 0} due to missing attribute")
                self.session.commit()


            if 'docs' in item:
                for doc in item['docs']:
                    doc_url = doc.get('doc_url', '')
                    if isinstance(doc_url, list) or isinstance(doc_url, set):
                        if len(doc_url) <= 1:
                            doc_url = next(iter(doc_url), None)
                    doc_obj = DocumentMedia(
                        product_id=product.product_id if hasattr(product, 'product_id') else 0,
                        url=doc_url,
                        type=doc.get('doc_type', ''),
                        description=doc.get('doc_desc', ''),
                        source=self.source,
                        ref_id=self.ref_id,  # Use the ref_id from impute_op
                    )
                    self.session.add(doc_obj)
                self.session.commit()


            if 'images' in item:
                for doc in item['images']:
                    doc_url = doc.get('image_url', '')
                    if isinstance(doc_url, list) or isinstance(doc_url, set):
                        if len(doc_url) <= 1:
                            doc_url = next(iter(doc_url), None)
                    doc_obj = DocumentMedia(
                        product_id=product.product_id if hasattr(product, 'product_id') else 0,
                        url=doc_url,
                        type='image',
                        source=self.source,
                        ref_id=self.ref_id,  # Use the ref_id from impute_op
                    )
                    self.session.add(doc_obj)
                self.session.commit()

            if 'compliance' in item:
                for ext in item['compliance']:
                    extra = ProductExtra(
                    product_id=product.product_id if hasattr(product, 'product_id') else 0,
                    name=ext.get('type', ''),
                    value=str(ext.get('value', '')),
                    source=self.source,
                    ref_id=self.ref_id,  # Use the ref_id from impute_op
                )
                self.session.add(extra)
                self.session.commit()                

            if 'extra' in item:
                for ext in item['extra']:
                    extra = ProductExtra(
                    product_id=product.product_id if hasattr(product, 'product_id') else 0,
                    name=ext.get('name', ''),
                    value=str(ext.get('value', '')),
                    source=self.source,
                    ref_id=self.ref_id,  # Use the ref_id from impute_op
                )
                self.session.add(extra)
                self.session.commit()                


            if 'sellers' in item:
                for seller in item['sellers']:
                    pseller = ProductSeller(
                    product_id=product.product_id if hasattr(product, 'product_id') else 0,
                    seller_name=seller.get('seller_name', ''),
                    seller_type=seller.get('seller_type', ''),
                    source=self.source,
                    ref_id=self.ref_id,  # Use the ref_id from impute_op
                )
                self.session.add(pseller)
                self.session.commit()  

            for key in  ['product_id', 'category', 'description', 'intro_date', 'country_of_origin', 'active_for_use', 
                         'manuf_standard_packaging', 'product_series', 'brand', 'title', 'lifecycle_status', 'package_type']:
                if key in item and item[key] is not None:
                    extra = ProductExtra(
                    product_id=product.product_id if hasattr(product, 'product_id') else 0,
                    name= key,
                    value=str(item[key]),
                    source=self.source,
                    ref_id=self.ref_id,  # Use the ref_id from impute_op
                )
            self.session.add(extra)
            self.session.commit()    


        except Exception as e:
            raise e
            return {"status": "error", "message": "Failed to write data to Supabase due to network issues or other errors."}
        print(f"Successfully wrote data to Supabase for {len(results)} products")
        return {"status": "success", "message": f"Data written to Supabase for {len(results)} products"}

```
